```python
import pandas as pd
from astropy.io import ascii
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**.dat to .csv Convertion**

```python
data_path = '../data/table1.dat'

colspecs = [
    (0, 19),    # ID
    (20, 28),   # RAdeg
    (29, 37),   # DEdeg
    (38, 43),   # e_RAdeg
    (44, 49),   # e_DEdeg
    (50, 58),   # RApeak
    (59, 67),   # DEpeak
    (68, 77),   # Sint
    (78, 85),   # e_Sint
    (86, 95),   # Speak
    (96, 103),  # e_Speak
    (104, 111), # rmspeak
    (112, 119), # e_rmspeak
    (120, 126), # thetamaj
    (127, 132), # thetamin
    (133, 139), # PA
    (140, 145), # alpha
    (146, 150), # e_alpha
    (151, 170), # ID2
    (171, 190), # ID3
]

columns = [
    "ID", "RAdeg", "DEdeg", "e_RAdeg", "e_DEdeg", "RApeak", "DEpeak", "Sint", "e_Sint",
    "Speak", "e_Speak", "rmspeak", "e_rmspeak", "thetamaj", "thetamin", "PA", "alpha",
    "e_alpha", "ID2", "ID3"
]


df = pd.read_fwf(data_path, colspecs=colspecs, names=columns)

for c in ["ID", "ID2", "ID3"]:
    df[c] = df[c].str.strip()

df.to_csv('../data/table1.csv', index=False)

```

```python
# Drop columns if they donâ€™t vary or are redundant (example: ID2, ID3)
df = df.drop(['ID2', 'ID3'], axis=1)
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['RAdeg'], df['DEdeg'], s=2, color='k')
plt.xlabel('Right Ascension (deg)')
plt.ylabel('Declination (deg)')
plt.title('Sky Map of VLA XMM-LSS/VIDEO Deep Field Radio Sources')
plt.gca().invert_xaxis()  # Astronomers plot RA decreasing to the right
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['RAdeg'], df['DEdeg'], 
            s=3, 
            c=df['Sint'], 
            cmap='plasma', 
            norm=plt.Normalize(vmin=df['Sint'].min(), vmax=df['Sint'].max()))
plt.xlabel('Right Ascension (deg)')
plt.ylabel('Declination (deg)')
plt.title('Sky Map (source brightness = color)')
plt.gca().invert_xaxis()
plt.grid(True)
plt.colorbar(label='Integrated Flux Density (mJy)')
plt.tight_layout()
plt.show()
```
